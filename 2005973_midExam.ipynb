```python
#4-1
data_14 = pd.read_csv('2014년_졸음운전교통사고.csv', encoding = 'euc-kr')
```

```python
data_15 = pd.read_csv('2015년_졸음운전교통사고.csv', encoding = 'euc-kr')
```

```python
data_16 = pd.read_csv('2016년_졸음운전교통사고.csv', encoding = 'euc-kr')
```

```python
data = pd.concat([data_14, data_15, data_16], 
                 ignore_index = True).set_index('구분')
data
```

```python
#4-2
data2 = data.copy()
```

```python
data2.index = pd.to_datetime(data2.index, format = '%Y년%m월')
data2.index
```

```python
data2.set_index([data2.index.year, data2.index.month], inplace = True)
```

```python
data2.index.names = ['년도', '월']
data2
```

```python
#4-3
pd.pivot_table (data = data2.reset_index(), index = '년도', values = '사망(명)',
               aggfunc = 'mean')
```

```python
pd.pivot_table (data = data2.reset_index(), index = '월', values = '사망(명)',
               aggfunc = 'mean')
```
